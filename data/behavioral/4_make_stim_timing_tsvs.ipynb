# Make timing files

make timing files for button press, labeled button press, video, and type labeled video

```python
# load template tsv
tsv = pd.read_table('template.tsv')

# for each subject 
for s in subs_1 + subs_2:

    # for each run / each logfile    
    for logfile in os.listdir(path+'/'+s):
        
        # skip non-log files
        if not logfile.endswith('.log'):
            continue
        
        # skip ignored logfiles
        if logfile in log_ignore_simple:
            continue
            
        # load the log file, get start time
        log = pd.read_table(path+'/'+s+'/'+logfile,header=None)
        exp_start = log[log[2]=='Keypress: equal'].iloc[0, 0]
        
        # Get the run number from the log line that imports the .csv file
        import_str = log[log[2].str.contains('.csv')].iloc[0, 2]
        run = int(re.search(r'movie_stim_(\d)\.csv', import_str).group(1))
        
        # skip exclude runs
        if s in set(exclude_df['sub']) and run in set(exclude_df[exclude_df['sub'] == s]['run']):
            print(f'excluding subject {s} run {run}')
        
        else:
            if s[-2] in ['1', '2']:
                sub_num = '0'+s[-2:]
            else:
                sub_num = '00'+s[-1:]
            

            ################ MAKE VIDEO EVENTS FILES ################

            ######## REAL V FAKE ########

            # tsv with videos (veridical vs scrambled)
            tsv_r_f = tsv.copy()

            # get all video start times
            raw_start = list(log[log[2].str.contains('movie: autoDraw = True')][0])
            start     = [x - exp_start for x in raw_start]
            
            
            # get all video types
            types = []
            for x in list(log[log[2].str.contains('.mp4')][2]):
                for t in ['veridical','scrambled']:
                    if t in x:
                        types.append(t)
                        break
                else:
                    types.append(pd.NA)
                    warnings.warn(f'video with data {x} is neither "veridical" nor "scrambled"')
                        
            tsv_r_f['onset'] = start
            tsv_r_f['trial_type'] = types
            # All videos are set to play for 3 seconds.
            tsv_r_f['duration'] = [3]*len(start)
            

            # save out to tsv
            filename = f'event_timing/sub-{sub_num}_ses-01_task-PredictingAttention_run-0{run}_labeled-type-videos.tsv'

            tsv_r_f.to_csv(filename, sep="\t", index=False)
            print(filename)
            
            
            ######## VIDEOS ########

            # tsv with videos (veridical vs scrambled)
            tsv_r_f = tsv.copy()

            # get all video start times
            raw_start = list(log[log[2].str.contains('movie: autoDraw = True')][0])
            start     = [x - exp_start for x in raw_start]
            

            tsv_r_f['onset'] = start
            tsv_r_f['trial_type'] = ['video']*len(start)
            tsv_r_f['duration'] = [3]*len(start)

            # save out to tsv
            filename = f'event_timing/sub-{sub_num}_ses-01_task-PredictingAttention_run-0{run}_videos.tsv'
            
            tsv_r_f.to_csv(filename, sep="\t", index=False)
            print(filename)
            
            
            ######## BUTTON_PRESSES ########

            # tsv with videos (veridical vs scrambled)
            tsv_r_f = tsv.copy()

            # get all video start times
            raw_start = list(log[(log[2].str.contains('Keypress: 1'))|(log[2].str.contains('Keypress: 2'))][0])
            start     = [x - exp_start for x in raw_start]

            tsv_r_f['onset'] = start
            tsv_r_f['trial_type'] = ['button']*len(start)
            tsv_r_f['duration'] = [.01]*len(start)

            # save out to tsv
            filename = f'event_timing/sub-{sub_num}_ses-01_task-PredictingAttention_run-0{run}_buttons.tsv'

            tsv_r_f.to_csv(filename, sep="\t", index=False)
            print(filename)
            
            
            ######## BUTTON_PRESSES 1 and 2 ########

            # tsv with videos (veridical vs scrambled)
            tsv_r_f = tsv.copy()

            # get all video start times
            raw_start = list(log[(log[2].str.contains('Keypress: 1'))|(log[2].str.contains('Keypress: 2'))][0])
            start     = [x - exp_start for x in raw_start]

            # get all video types
            types = []
            for x in list(log[(log[2].str.contains('Keypress: 1'))|(log[2].str.contains('Keypress: 2'))][2]):
                for t,lab in zip(['1','2'],['one','two']):
                    if t in x:
                        types.append(lab)
                        break
                else:
                    types.append(pd.NA)
                    warnings.warn(f'keypress {x} contains neither "1" nor "2"')

            tsv_r_f['onset'] = start
            tsv_r_f['trial_type'] = types
            tsv_r_f['duration'] = [.01]*len(start)

            # save out to tsv
            filename = f'event_timing/sub-{sub_num}_ses-01_task-PredictingAttention_run-0{run}_labeled-type-buttons.tsv'

            tsv_r_f.to_csv(filename, sep="\t", index=False)
            print(filename)
            
```

### make video judge label timing files

```python
# loop through every existing tsv video file
for t in os.listdir('event_timing/'):
    # skip other files
    if not t.endswith('_videos.tsv'):
        continue
    
    # obtain the subject and run # from filename
    run = re.search('run-(\d+)', t).group(1)
    sub = re.search('sub-(\d+)', t).group(1)
    
    sub_short = sub.lstrip("0")

        
    # load the corresponding .csv file
    csv_files = [ x for x in os.listdir(f'{path}/subject_{sub_short}') if 'csv' in x and x not in csv_ignore_simple ]
    csv_data  = [pd.read_csv(f'{path}/subject_{sub_short}/{x}') for x in csv_files]
    # note: we won't worry about 'exclude' or an 'ignore' because we didn't make earlier tsv files for those
    #       none of the bad ones will match our existing files
    
    
    for cd, cf in zip(csv_data,csv_files):
        if cd.shape[0] > 1 and cd.loc[1]['run'] == int(run):
            df = cd[~cd['run'].isna()].copy()
                
            df['resp'] = df['key_resp_9.keys'].str.split("'",expand=True)[1]
            
            # Dictionary to map integer values to words
            word_dict = {'1': "one", '2': "two", 'None': None}
            # Replace values in 'col1' using the dictionary
            df['resp'] = df['resp'].replace(word_dict)
            
            # load the tsv as table
            table = pd.read_table('event_timing/'+t)
            
            # map the responses onto the table and remove non-response rows
            table['trial_type'] = list(df['resp'])
            table = table[table['trial_type'].isin(['one','two'])]

            # output the new table as tsv
            filename = t.replace('videos.tsv','labeled-judge-videos.tsv')
            filename = 'event_timing/'+filename
            table.to_csv(filename, sep='\t', index=False)
            print(filename)
```
