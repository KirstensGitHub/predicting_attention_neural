### Check timing

#### Function for getting the number of frames in a video

```python
def get_frames(l):
    '''
    input:  list of movie filepaths
    output: list of # of frames
    '''

    frames = []

    for path in l:
        
        # Define the command
        command = ["ffprobe", "-v", "error", "-select_streams", "v:0", "-count_frames", "-show_entries", "stream=nb_read_frames", "-print_format", "csv", path]
        
        # Execute the command and capture output
        output = subprocess.run(command, capture_output=True).stdout.decode("utf-8")
        
        # Extract the number of frames using regex
        pattern = r"(\d+)"
        match = re.search(pattern, output)
        
        if match:
            number_of_frames = int(match.group(1))
        else:
            number_of_frames = np.nan
            print('could not find number of frames: '+path)

        frames.append(number_of_frames)
    
    return(frames)
            
```

#### Get the duration of each run

```python
real_timing_dict = {1:{'subject':[],'duration':[], 'run':[]}, 2:{'subject':[],'duration':[], 'run':[]}, 3:{'subject':[],'duration':[], 'run':[]}, 
                    4:{'subject':[],'duration':[], 'run':[]}, 5:{'subject':[],'duration':[], 'run':[]}, 6:{'subject':[],'duration':[], 'run':[]}}

subject_dirs = [x for x in os.listdir('.') if 'ubject' in x]

# for each person
for s in subject_dirs:

    # for each run / each logfile
    #print(len([y for y in os.listdir(s) if y[-4:]=='.log']))
    for logfile in [y for y in os.listdir(s) if y[-4:]=='.log']:

        # load the log file and get run #
        log = pd.read_table(s+'/'+logfile,header=None)

        # provided that there were scan pulses recorded
        if log[log[2]=='Keypress: equal'].shape[0]>0:

            # get run #
            run = list(log[log[2].str.contains('.csv')][2].str.split(' '))[0][1][-5]
            
            # find onset of first fixation and end of last fixation
            exp_start = log[log[2]=='Keypress: equal'].iloc[0][0]
            exp_end = log[log[2]=='fix4: autoDraw = False'][0].unique()[0]
    
            # run duration
            duration = exp_end - exp_start

        # else:
        #     duration = np.nan; run = np.nan
    
            # add the total time to dict, in the key for that run (1-6)
            real_timing_dict[int(run)]['subject' ].append(s)
            real_timing_dict[int(run)]['duration'].append(duration)
            real_timing_dict[int(run)]['run'].append(run)
    
```

```python
# get theoretical timings

theo_timing_dict = {1:{}, 2:{}, 3:{}, 4:{}, 5:{}, 6:{}}
example_subject  = 'Subject_23'

# KZ NOTE: swap out subject 23 with the csv stim files from the MRI room computer
# KZ NOTE: swap out the fps 60 stimuli with the stimuli from the MRI room computer

# for each run
for file in [x for x in os.listdir(example_subject) if 'csv' in x]:

    # load the csv
    df = pd.read_csv(example_subject+'/'+file); run = int(df['run'][1])

    # get the total jitter durations
    theo_timing_dict[run]['jitters'] = list(df[1:21]['jitters']) 

    # get the names of the videos in that run
    theo_timing_dict[run]['videos'] = list(df[1:21]['moviestim']) 
    
    # get the number of frames in each video
    theo_timing_dict[run]['movie_frames'] = get_frames(theo_timing_dict[run]['videos'])

    # video_time = total_frames * (1/60) <-- 60 frames / second
    theo_timing_dict[run]['video_times'] = [x*(1/60) for x in theo_timing_dict[run]['movie_frames'] ]
    
    # all_other_time = ( 20 response_periods * 3 secs ) + 10s post_task_fixation
    theo_timing_dict[run]['response_periods'] = 20*3
    theo_timing_dict[run]['post_fixation']    = 10
    
    # theo_timing_dict[run]['duration'] = sum(video_times) + sum(response_periods) + post_fixation 
    theo_timing_dict[run]['duration'] = np.sum(theo_timing_dict[run]['video_times']) + np.sum(theo_timing_dict[run]['jitters']) + theo_timing_dict[run]['response_periods'] + theo_timing_dict[run]['post_fixation']
```
